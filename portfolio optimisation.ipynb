```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# Download historical data for Tata Motors from Yahoo Finance
tata_motors_data = yf.download("TATAMOTORS.BO", start="2020-04-04", end="2024-04-04")

# Calculate daily returns
tata_motors_data['Returns'] = tata_motors_data['Adj Close'].pct_change()

# Drop NaN values (first row)
returns = tata_motors_data['Returns'].dropna()

# Calculate average risk and average return for the past 5 years
average_risk = np.std(returns)
average_return = np.mean(returns)

# Annualize risk and return
trading_days_per_year = 252
annual_risk = average_risk * np.sqrt(trading_days_per_year) * 100
annual_return = average_return * trading_days_per_year * 100

print("Annualized Risk (Standard Deviation):", annual_risk, "%")
print("Annualized Return:", annual_return, "%")
```

```python
# Plot
plt.figure(figsize=(8, 6))
plt.scatter(annual_risk, annual_return, color='blue')
plt.title('Annualized Risk vs. Return for Tata Motors (2020-2024)')
plt.xlabel('Annualized Risk (%)')
plt.ylabel('Annualized Return (%)')
plt.grid(True)
plt.show()
```

```python
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt

# Define function to fetch data and calculate annualized risk and return
def fetch_data(symbol, start_date, end_date):
    data = yf.download(symbol, start=start_date, end=end_date)
    data['Returns'] = data['Adj Close'].pct_change()
    returns = data['Returns'].dropna()
    average_risk = np.std(returns) * np.sqrt(252) * 100
    average_return = np.mean(returns) * 252 * 100
    return average_risk, average_return

# Define symbols and date ranges for each investment instrument
instruments = [
    {"symbol": "TATAMOTORS.BO", "start": "2019-04-04", "end": "2024-04-04", "color": "red", "label": "Tata Motors"},
    {"symbol": "ICICIPRULI.NS", "start": "2019-04-04", "end": "2024-04-04", "color": "magenta", "label": "ICICI Prudential Value Discovery Fund"},
    {"symbol": "KOTAKGOLD.NS", "start": "2020-01-01", "end": "2023-04-04", "color": "gold", "label": "Kotak Gold ETF"},
    {"symbol": "EMBASSY.NS", "start": "2020-01-01", "end": "2023-04-04", "color": "brown", "label": "Embassy Office Parks REIT"},
    {"symbol": "^NSEI", "start": "2020-01-01", "end": "2023-04-04", "color": "purple", "label": "Indian Govt. Securities 10Y Bond"}
]

# Fetch data and calculate annualized risk and return for each instrument
risk_return_data = []
for instrument in instruments:
    risk, ret = fetch_data(instrument["symbol"], instrument["start"], instrument["end"])
    risk_return_data.append({"risk": risk, "return": ret, "color": instrument["color"], "label": instrument["label"]})

# Plot
plt.figure(figsize=(10, 8))
for data in risk_return_data:
    plt.scatter(data["risk"], data["return"], color=data["color"], label=data["label"])

plt.title('Annualized Risk vs. Return')
plt.xlabel('Annualized Risk (%)')
plt.ylabel('Annualized Return (%)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Prepare data for optimization
risk_data = np.array([data["risk"] for data in risk_return_data])
return_data = np.array([data["return"] for data in risk_return_data])

# Define optimization parameters
num_assets = len(instruments)
num_portfolios = 10000

# Generate random portfolios with different weights
np.random.seed(42)  # for reproducibility
all_weights = np.zeros((num_portfolios, num_assets))
ret_arr = np.zeros(num_portfolios)
vol_arr = np.zeros(num_portfolios)
sharpe_arr = np.zeros(num_portfolios)

for i in range(num_portfolios):
    # Generate random weights
    weights = np.random.random(num_assets)
    weights /= np.sum(weights)

    # Save weights
    all_weights[i, :] = weights

    # Calculate portfolio return and volatility
    ret_arr[i] = np.sum(return_data * weights)
    vol_arr[i] = np.sqrt(np.dot(weights.T, np.dot(np.cov(return_data, rowvar=False), weights)))

    # Calculate Sharpe ratio
    sharpe_arr[i] = ret_arr[i] / vol_arr[i]

# Find portfolio with maximum Sharpe ratio
max_sharpe_idx = np.argmax(sharpe_arr)
max_sharpe_return = ret_arr[max_sharpe_idx]
max_sharpe_volatility = vol_arr[max_sharpe_idx]
max_sharpe_weights = all_weights[max_sharpe_idx, :]

# Plot efficient frontier
plt.figure(figsize=(10, 8))
plt.scatter(vol_arr, ret_arr, c=sharpe_arr, cmap='viridis')
plt.colorbar(label='Sharpe Ratio')
plt.scatter(max_sharpe_volatility, max_sharpe_return, marker='*', color='red', s=100, label='Max Sharpe Ratio')
plt.title('Efficient Frontier')
plt.xlabel('Volatility')
plt.ylabel('Return')
plt.legend()
plt.grid(True)
plt.show()

# Print optimized portfolio weights and metrics
print("Optimized Portfolio Weights:")
for instrument, weight in zip(instruments, max_sharpe_weights):
    print(f"{instrument['label']}: {weight:.2f}")

print("\nOptimized Portfolio Return:", max_sharpe_return)
print("Optimized Portfolio Volatility:", max_sharpe_volatility)
```

```python
# Prepare data for optimization
risk_data = np.array([data["risk"] for data in risk_return_data])
return_data = np.array([data["return"] for data in risk_return_data])

# Add risk-free asset
risk_free_rate = 1.5 / 100  # Risk-free rate assumed to be 1.5%
risk_data = np.append(risk_data, 0)  # Risk of risk-free asset is zero
return_data = np.append(return_data, risk_free_rate)  # Return of risk-free asset is risk-free rate

# Define optimization parameters
num_assets = len(instruments) + 1  # Adding one for the risk-free asset
num_portfolios = 10000

# Generate random portfolios with different weights
np.random.seed(42)  # for reproducibility
all_weights = np.zeros((num_portfolios, num_assets))
ret_arr = np.zeros(num_portfolios)
vol_arr = np.zeros(num_portfolios)
sharpe_arr = np.zeros(num_portfolios)

for i in range(num_portfolios):
    # Generate random weights
    weights = np.random.random(num_assets)
    weights /= np.sum(weights)

    # Save weights
    all_weights[i, :] = weights

    # Calculate portfolio return and volatility
    ret_arr[i] = np.sum(return_data * weights)
    vol_arr[i] = np.sqrt(np.dot(weights.T, np.dot(np.cov(return_data, rowvar=False), weights)))

    # Calculate Sharpe ratio
    sharpe_arr[i] = (ret_arr[i] - risk_free_rate) / vol_arr[i]

# Find portfolio with maximum Sharpe ratio
max_sharpe_idx = np.argmax(sharpe_arr)
max_sharpe_return = ret_arr[max_sharpe_idx]
max_sharpe_volatility = vol_arr[max_sharpe_idx]
max_sharpe_weights = all_weights[max_sharpe_idx, :]

# Plot efficient frontier
plt.figure(figsize=(10, 8))
plt.scatter(vol_arr, ret_arr, c=sharpe_arr, cmap='viridis')
plt.colorbar(label='Sharpe Ratio')
plt.scatter(max_sharpe_volatility, max_sharpe_return, marker='*', color='red', s=100, label='Max Sharpe Ratio')
plt.title('Efficient Frontier (with Risk-Free Asset)')
plt.xlabel('Volatility')
plt.ylabel('Return')
plt.legend()
plt.grid(True)
plt.show()

# Print optimized portfolio weights and metrics
print("Optimized Portfolio Weights:")
for instrument, weight in zip(instruments, max_sharpe_weights[:-1]):  # Exclude last weight for risk-free asset
    print(f"{instrument['label']}: {weight:.2f}")
print("Risk-Free Asset: ", max_sharpe_weights[-1])  # Print weight for risk-free asset

print("\nOptimized Portfolio Return:", max_sharpe_return)
print("Optimized Portfolio Volatility:", max_sharpe_volatility)
```

```python
# We can see that including the risk free asset althought the the return has reduced a little bit along with the risk. 
#So it is a good idea to reduced the risk while maintaining a good return rate.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf

# Calculate portfolio returns using optimized weights
portfolio_returns = np.dot(max_sharpe_weights[:-1], return_data[:-1]) + max_sharpe_weights[-1] * risk_free_rate

# Convert the portfolio returns to a pandas DataFrame with a datetime index
dates = pd.date_range(start='2019-04-04', end='2024-04-04', freq='B')  # Assuming daily frequency
portfolio_returns_df = pd.DataFrame(data=portfolio_returns, index=dates, columns=['Portfolio Returns'])

# Plot ACF for portfolio returns with custom parameters
plt.figure(figsize=(12, 6))
plot_acf(portfolio_returns_df['Portfolio Returns'], lags=40, ax=plt.gca(), zero=False)
plt.title('Autocorrelation Function (ACF) for Optimized Portfolio Returns')
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.grid(True)
plt.show()
```
